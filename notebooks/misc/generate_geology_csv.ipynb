```python
import numpy as np
```

```python
save_path = "../examples/inputs/geology.csv"
```

```python
def generate_random_geology(nx, ny, n):
    return np.rint(np.random.rand(nx,ny) * n)
```

```python
topo = generate_random_geology(300,300, 2)
```

```python
np.savetxt(save_path, topo, delimiter=",")
```
