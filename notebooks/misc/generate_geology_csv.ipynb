```python
import numpy as np
```

```python
save_path = "../examples/inputs/geology.csv"
```

```python
def generate_random_geology(nx, ny, n):
    return np.rint(np.random.rand(nx,ny) * n)

def generate_linear_geology(nx, ny, n):
    d = np.zeros((nx,ny))
    for y in range(ny):
        for x in range(nx):
            d[x,y] = int((x/nx) * n)
    return d
```

```python
topo = generate_linear_geology(300,300, 10)
```

```python
np.savetxt(save_path, topo, delimiter=",")
```
