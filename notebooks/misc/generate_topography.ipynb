```python
import numpy as np
```

```python
save_path = "../examples/inputs/topography.csv"
```

```python
def generate_random_topography(nx, ny):
    return np.rint(np.random.rand(nx,ny))

def generate_slopy_topography(nx, ny):
    d = np.zeros((nx,ny))
    for y in range(ny):
        for x in range(nx):
            d[x][y] = x*y
    return d
```

```python
topo = generate_slopy_topography(300,300)
```

```python
np.savetxt(save_path, topo, delimiter=",")
```
