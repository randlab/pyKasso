A simple example
================

To illustrate how *pyKasso* can be used, let us start with a simple example. <br>
The aim is to generate a stochastic karstic network from scratch.


Let's start!
------------

First, we have to import the *pyKasso* module:

```python
import pykasso as pk
```

Then, we create an instance of the `SKS` class. If no argument is given, *pyKasso* will use the default settings file. <br>
Here, in the dedault configuration, `rand_seed` equals 1, it means the seed will be always the same. It could be used in order to guarantee the reproducibility of the example:

```python
catchment = pk.SKS()
catchment.set_verbosity(4)
```

Visualize the generated environment
-----------------------------------

Once the ``SKS`` class is constructed, it contains now all the geological data.
With this default configuration, only the fractures and inlets positions are stochastically generated. <br>
We can access to the geological data with the ``get`` functions, or we can print it out with the ``show_cacthment()`` method.

```python
# Return a numpy-array
geology     = catchment.get_geology()
topography  = catchment.get_topography()
orientation = catchment.get_orientation()
faults      = catchment.get_faults()
fractures   = catchment.get_fractures()

# Print out directly the requested data
catchment.show_catchment(data='geology')
catchment.show_catchment(data='faults')
catchment.show_catchment(data='fractures')
```

Compute a karst network
-----------------------

To start a stochastic karst network simulation, simply use the ``compute_karst_network()`` method:

```python
catchment.compute_karst_network()
```

The print out above come from the *karstnet* dependance and is, at this stage, not relevant. <br>
Once the calculations are over, the results are stored in a ``KarstNetwork`` object then piled-up in a ``SKS`` list. <br>
A ``KarstNetwork`` object store several useful data from a simulation in different attributes:
- ``maps`` : the maps calculated by *pyKasso*
- ``points`` : the positions of the inlets and outlets
- ``network`` : the positions of each node from the network and the associated bounds
- ``stats`` : the statistics calculated by *karstnet*
- ``settings`` : the settings used from the simulation

```python
catchment.karst_simulations
```
