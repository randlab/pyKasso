# KarstNetwork class
This notebook display how the KarstNetwork() class works. 

```python
import pykasso as pk
```

```python
catchment = pk.SKS()
```

```python
catchment.compute_karst_network()
```

```python
# After karst computing, pykasso stores the result in a KarstNetwork instance and add it in a list named  'karst_simulations'
catchment.karst_simulations
```

```python
catchment.karst_simulations[0]
```

```python
# The KarstNetwork is an object with 5 entry :
# - maps     -  
# - points   - 
# - network  - 
# - stats    - 
# - settings - 
catchment.karst_simulations[0].maps
catchment.karst_simulations[0].points
catchment.karst_simulations[0].network
catchment.karst_simulations[0].stats
catchment.karst_simulations[0].settings
```
