# Polygon class
This notebook display how the Polygon() class works. 

```python
import pykasso as pk
```

```python
# First create grid object
x0, y0 ,z0 = 0, 0, 0
nx, ny, nz = 6, 6, 1
dx, dy, dz = 1, 1, 1

grid = pk.Grid(x0,y0,z0,nx,ny,nz,dx,dy,dz)
```

```python
# Then create polygon manager
polygon = pk.Polygon(grid)
```

```python
# Set a polygon with a list of coordinates [[x1,y1],[x2,y2],[xn,yn]]
# Inspect if vertices are well inside the grid
# Prints something only when at least one vertex is outside the grid limits 
# sys.exit() if no vertex is in the grid limits
polygon.set_polygon([[3,3],[3,7],[7,7],[7,5]])
```

```python
# 'polygon' attribute is a Path instance from matplotlib
# Vertices are strored in the 'vertices' attribute
print(polygon.polygon)
print('vertices :', polygon.vertices)
```

```python
# Display polygon and grid limits
f = polygon.show()
```
