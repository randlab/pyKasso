# Fracture class
This notebook display how the Fracture() class works. 

```python
import pykasso as pk
```

```python
# Fracture instances should be created and managed with the GeologyManager class.
ID = 0
family = 1
position = [0,0,0]
radius = 5
orientation = 50
dip = 50
normal = [5,4,3]

fracture = pk.Fracture(ID, 
                       family, 
                       position, 
                       radius,
                       orientation,
                       dip, 
                       normal)
```

```python
# Print the fracture
print(fracture)
```

```python
# Attributes can be easily retrieved with approprieted methods
f_id             = fracture.get_ID()
f_family         = fracture.get_family()
f_x, f_y, f_z    = fracture.get_position()
f_radius         = fracture.get_radius()
f_orientation    = fracture.get_orientation()
f_dip            = fracture.get_dip()
f_na, f_nb, f_nc = fracture.get_normal()

print("Fracture n°", f_id)
print("Family : ", f_family)
print("Position : ({},{},{})".format(f_x, f_y, f_z))
```
