# Grid class
This notebook display how the Grid() class works. 

```python
import pykasso as pk
```

```python
# Grid parameters
x0 = 0
y0 = 0
z0 = 0
nx = 3
ny = 3
nz = 3
dx = 1
dy = 2
dz = 3
```

```python
# Create a grid
grid = pk.Grid(x0,y0,z0,nx,ny,nz,dx,dy,dz)
print(grid)
```

```python
# Get area and volume
print("area   :", grid.area)
print("volume :", grid.volume)
```

```python
# Get limits
print("x limits : ", grid.xlimits)
print("y limits : ", grid.ylimits)
print("z limits : ", grid.zlimits)
```

```python
# Get extent [xlimits[0], xlimits[1], ylimits[0], ylimits[1]]
print("extent : ", grid.extent)
```

```python
# Get node positions
print("x : ", grid.x)
print("y : ", grid.y)
print("z : ", grid.z)
```

```python
# Get meshgrids
print("X meshgrid : \n", grid.X)
print("Y meshgrid : \n", grid.Y)
print("Z meshgrid : \n", grid.Z)
```

```python
# Get i,j and k with x, y and z-coordinate
# Return index even if x,y,z is outside the grid
print("i : ", grid.get_i(-1))
print("j : ", grid.get_j(0))
print("k : ", grid.get_k(7.5))
```

```python
# Get x, y and z-coordinate with i,j and k index 
# Return none if index is outside the grid
print("x : ", grid.get_x(0))
print("y : ", grid.get_y(2))
print("z : ", grid.get_z(10))
```

```python
# Check if a point (x,y,z) is inside the grid
grid.is_inbox(1,1,1)
```
