```python
import pykasso as pk
```

```python
# First create grid object
x0, y0 ,z0 = 0, 0, 0
nx, ny, nz = 6, 6, 6
dx, dy, dz = 1, 1, 1

grid = pk.Grid(x0,y0,z0,nx,ny,nz,dx,dy,dz)
```

```python
# Then create polygon manager
polygon = pk.Polygon(grid)
```

```python
# Set a polygon with a list of coordinates [[x1,y1],[x2,y2],[xn,yn]]
polygon.set_polygon([[3,3],[3,7],[7,7],[7,5]])
```

```python
# Retrieve polygon vertices coordinates
print(polygon.polygon)
```

```python
# Inspect if vertices are well inside the grid
# Could be useful when importing numberous vertices
# Return something only when at least one vertice is outside the grid limits 
polygon.inspect_polygon()
```

```python
# Display polygon and grid limits
polygon.show()
```

```python
# Get the mask
polygon.mask
```

```python
# Clean the polygon instance
polygon.clean_polygon()
print(polygon.polygon)
print(polygon.mask)
```
